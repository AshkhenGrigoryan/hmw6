```python
import quandl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open ("key.txt",'r') as k:
    key1 = k.read()
data = quandl.get(['WORLDBANK/ARM_TOT','WORLDBANK/LUX_TOT'], authtoken = key1)

data.plot()
plt.xlabel("Years")
plt.ylabel("Terms of Trade")
plt.show()
```

```python
data=quandl.get("WORLDBANK/ARM_TOT", authtoken=key1)
data1 = data.sort_values(by='Value', ascending=0)
max2=data1['Value'][1]
print "2nd highest value for Armenia is  " + str(max2)
```

```python
exp = quandl.get("WTO/MERCH_EXP_ARM", authtoken=key1)
imp = quandl.get("WTO/MERCH_IMP_ARM", authtoken=key1)
Net_export = exp ["Value"]-imp["Value"]
Net_export.plot()
plt.ylabel("Net_Export")
```

```python
import googlemaps
with open ("gapi.txt",'r') as f:
    key2 = f.read()
gmaps=googlemaps.Client(key2)
dist =[]
cities = ["Washington, DC","Los Angeles, CA", "Las Vegas, NV","Chicago, IL", "San Francisco, CA", "Dallas, TX","Garden City, KS", "Fort Collins, CO", "Miami, FL","Kansas City, MO"]

for i in cities:
        directions=gmaps.directions(i,"New York, NY", units = "metric")
        dist.append(directions[0]["legs"][0]["distance"])
        
for i in range(0,len(dist)):
    print('The distance between ' +  cities[i] + ' and ' + " New York, NY" + ' is ' + dist[i]["text"])
```
